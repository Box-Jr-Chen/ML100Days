### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)


```

```python
df = pd.DataFrame(data)
```

```python
countLIST = []
for i in range(df.shape[0]):
    count = df['postCount'][i]['last30Days']
    countLIST.append(count)
```

```python

df['postCount'] = countLIST
```

```python
df.sort_values(by = 'postCount', ascending=False )
```

```python
df.to_csv(r'Day_014_HW_20210401.csv', index = False)
```
