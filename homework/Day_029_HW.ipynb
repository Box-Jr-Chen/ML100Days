```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## 輸入資料
boys=np.array([164, 176, 169, 169, 165, 175, 159, 151, 144, 160, 183, 165, 156, 170,
 164, 173, 165, 163, 177, 171])


girls=np.array([169, 170, 162, 154, 183, 173, 169, 167, 170, 185, 162, 175, 168, 151,
 181, 170, 182, 156, 159, 160])
```

```python
#Q1:試著用今天所教的內容，如何描述這兩組資料的樣態?
#暗示:  分別計算男生和女生的統計量
mean_boy=np.mean(boys)
print("男孩身高平均=",mean_boy)
mean_girl=np.mean(girls)
print("女孩身高平均=",mean_girl)


statistics_mean_boy=statistics.mean(boys)
print("statistics_mean_boy=",statistics_mean_boy)
statistics_mean_girls=statistics.mean(girls)
print("statistics_mean_girls=",statistics_mean_girls)

print("================================")

# 計算統計量_中位數的方法
np_median_boy=np.median(boys,axis=None)
print("男孩身高中位數=",np_median_boy)

statistics_median_boy=statistics.median(boys)
print("statistics_median_boy=",statistics_median_boy)

np_median_girl=np.median(girls,axis=None)
print("女孩身高中位數=",np_median_girl)

statistics_median_girl=statistics.median(girls)
print("statistics_median_boy=",statistics_median_girl)

print("================================")
# 統計量_眾數
# 統計量的眾數，如果有多個眾數，取最小的值當眾數。

mode_boy=stats.mode(boys,axis=None)
print("男孩身高眾數=",mode_boy)
print("男孩身高眾數=",mode_boy[0][0])


# 統計量_眾數
statistics_mode_boys=statistics.mode(boys)
print("statistics_mode_boy=",statistics_mode_boys)


mode_girls=stats.mode(girls,axis=None)
print("女孩身高眾數=",mode_girls)
print("女孩身高眾數=",mode_girls[0][0])


# 統計量_眾數
statistics_mode_girls=statistics.mode(girls)
print("statistics_mode_girls=",statistics_mode_girls)

print("================================")

#全距
#rangeV=max(boys)-min(boys)
def rangeV(x): 
  return(max(x)-min(x))
    
print("男生全距"+ str(rangeV(boys)))
print("女生全距"+ str(rangeV(girls)))

print("================================")
# 計算變異數的方法
print("男孩身高變異數=",statistics.variance(boys))
print("男孩身高變異數=",np.var(boys,ddof=1))


print("女孩身高變異數=",statistics.variance(girls))
print("女孩身高變異數=",np.var(girls,ddof=1))


print("================================")
# 統計量_標準差的方法
#樣本標準差
#ddof=1, 回傳 sample standard deviation 樣本標準差，分母(n-1)，無偏估計
std_boy=np.std(boys,ddof=1)
print("男孩身高標準差=",std_boy)

statistics_stdev_boy=statistics.stdev(boys)
print("statistics_mean_boy=",statistics_stdev_boy)


std_girls=np.std(girls,ddof=1)
print("女孩身高標準差=",std_girls)

statistics_stdev_girls=statistics.stdev(girls)
print("statistics_mean_girls=",statistics_stdev_girls)

print("================================")
# python 百分位數
#np
print("男孩90百分位數=",np.percentile(boys, 90))
print("50百分位數=",np.percentile(boys, 50))
print("20百分位數=",np.percentile(boys, 20))
#stat
print("20百分位數=",stats.scoreatpercentile(boys, 20))

#np
print("女孩90百分位數=",np.percentile(girls, 90))
print("50百分位數=",np.percentile(girls, 50))
print("20百分位數=",np.percentile(girls, 20))
#stat
print("20百分位數=",stats.scoreatpercentile(girls, 20))

print("================================")
print("男孩峰度=",stats.skew(boys))
print("男孩偏度=",stats.kurtosis(boys))

print("女孩峰度=",stats.skew(girls))
print("女孩偏度=",stats.kurtosis(girls))
print("================================")

#最後，畫圖看分布
plt.hist(boys,alpha=.4,bins=40)
plt.title('boy,skewness={0},kurtosis={1}'.format(round(stats.skew(boys),2),round(stats.kurtosis(boys),2)))
plt.axvline(x=mean_boy)
plt.show()

plt.hist(boys,alpha=.4,bins=40)
plt.title('girl,skewness={0},kurtosis={1}'.format(round(stats.skew(girls),2),round(stats.kurtosis(girls),2)))
plt.axvline(x=mean_girl)
plt.show()
```
