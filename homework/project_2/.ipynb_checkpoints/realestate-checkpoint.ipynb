# 4. 找出台北市時價登入總價高度相關的變數

阿宏我是台北人他想找出影響台北市總價、單價元坪的因子<br>
相關係數0.3以下為低相關，0.3\~0.7為中等相關，0.7以上為高度相關<br>
1. 資料切片切出city欄位為台北市的資料，並找出時價登入總價(total_price)高度相關的變數
2. 資料切片切出city欄位為台北市的資料，找出單價元坪(unit_price_square_feet)高度相關的變數

```python
#1. 資料切片切出city欄位為台北市的資料，
df_main_use_Taipei = df_main_use[df_main_use['city'] == '台北市']
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import researchpy
import math
import pingouin as pg
import seaborn as sb
```

```python
# 1. 資料切片切出city欄位為台北市的資料，並找出時價登入總價(total_price)高度相關的變數
# 相關係數0.3以下為低相關，0.3\~0.7為中等相關，0.7以上為高度相關

for o in df_main_use_Taipei:
    if o != 'total_price':
        if df_main_use_Taipei[o].dtype == 'int64' or  df_main_use_Taipei[o].dtype == 'int32': # 離散 VS 離散 Cramer's V
            
            contTable = pd.crosstab(df_main_use_Taipei['total_price'], df_main_use_Taipei[o])
            df        = min(contTable.shape[0], contTable.shape[1])-1
            crosstab, res = researchpy.crosstab(df_main_use_Taipei['total_price'], df_main_use_Taipei[o], test='chi-square')
            
            if math.isnan(res.loc[2,'results']) ==False :
                if  res.loc[2,'results'] < 0.3 :
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 低相關")
                elif  res.loc[2,'results'] >= 0.3  and  res.loc[2,'results'] < 0.7:
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 中相關")
                elif  res.loc[2,'results'] >= 0.7 :
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 高相關")
                        
        elif df_main_use_Taipei[o].dtype == 'float64': #離散 VS 連續 Eta Squared
            
            aov   = pg.anova(dv='total_price', between=o, data=df_main_use_Taipei, detailed=True)
            etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
            
            if math.isnan(etaSq) ==False :
                if  etaSq < 0.3 :
                        print(" Eta Squared :"+ str(etaSq) +"; "+o +" : 低相關")
                elif  etaSq >= 0.3  and  etaSq < 0.7:
                        print(" Eta Squared :"+str(etaSq) +"; "+o +" : 中相關")
                elif  etaSq >= 0.7 :
                        print(" Eta Squared :"+ str(etaSq) +"; "+o +" : 高相關")

```

```python
# 2. 資料切片切出city欄位為台北市的資料，找出單價元坪(unit_price_square_feet)高度相關的變數
# 相關係數0.3以下為低相關，0.3\~0.7為中等相關，0.7以上為高度相關


for o in df_main_use_Taipei:
    if o != 'unit_price_square_feet':
        if df_main_use_Taipei[o].dtype == 'int64' or  df_main_use_Taipei[o].dtype == 'int32': # 離散 VS 離散 Cramer's V
            
            contTable = pd.crosstab(df_main_use_Taipei['unit_price_square_feet'], df_main_use_Taipei[o])
            df        = min(contTable.shape[0], contTable.shape[1])-1
            crosstab, res = researchpy.crosstab(df_main_use_Taipei['unit_price_square_feet'], df_main_use_Taipei[o], test='chi-square')
            
            if math.isnan(res.loc[2,'results']) ==False :
                if  res.loc[2,'results'] < 0.3 :
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 低相關")
                elif  res.loc[2,'results'] >= 0.3  and  res.loc[2,'results'] < 0.7:
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 中相關")
                elif  res.loc[2,'results'] >= 0.7 :
                        print(" Cramer's V :"+ str(res.loc[2,'results']) +"; "+ o +" : 高相關")
                        
        elif df_main_use_Taipei[o].dtype == 'float64': #離散 VS 連續 Eta Squared
            
            aov   = pg.anova(dv='unit_price_square_feet', between=o, data=df_main_use_Taipei, detailed=True)
            etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
            
            if math.isnan(etaSq) ==False :
                if  etaSq < 0.3 :
                        print(" Eta Squared :"+ str(etaSq) +"; "+o +" : 低相關")
                elif  etaSq >= 0.3  and  etaSq < 0.7:
                        print(" Eta Squared :"+str(etaSq) +"; "+o +" : 中相關")
                elif  etaSq >= 0.7 :
                        print(" Eta Squared :"+ str(etaSq) +"; "+o +" : 高相關")
```
