chinese_score = np.array([65,90,82,72,66,77])

上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，運用上列數據回答下列問題。



有多少學生英文成績比數學成績高?
是否全班同學最高分都是國文?

```python
english_score = np.array([55,89,76,65,48,70])

math_score = np.array([60,85,60,68,55,60])

chinese_score = np.array([65,90,82,72,66,77])
```

```python
dt = np.dtype({'names':('english','math','chinese'),'formats':(np.uint8,np.uint8,np.uint8)})
```

```python
c = np.zeros(6,dtype=dt)
```

```python
c['english'] =english_score
c['math'] =math_score
c['chinese'] =chinese_score
```

1.答案

```python
c[c['english'] > c['math']].size #答案
```

2.答案

```python
(np.max(c['chinese']) >  np.max(c['math'])) and (np.max(c['chinese']) >  np.max(c['english'])) 
```
