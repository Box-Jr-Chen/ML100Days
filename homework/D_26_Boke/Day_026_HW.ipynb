# 作業: 處理分類數據

1.建立簡單的水果資料集 fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'] counts = [5, 3, 4, 2, 4, 6]

2.利用vbar 水果名稱因子列表作為x 坐標，將條形高度作為top坐標，以及可選地width我們想要設置的任何屬性進行調用：

p.vbar(x=fruits, top=[5, 3, 4, 2, 4, 6], width=0.9)

由於Bokeh會按順序顯示條形，因此在條形圖中“排序”條形與對范圍的因子進行排序相同。

sorted_fruits = sorted(fruits, key=lambda x: counts[fruits.index(x)])

```python
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure

#指定輸出檔名
output_file("fruit_info.html")

#給定資料
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]


ColumnDataSource(data=dict(fruits=fruits, counts=counts, color=Spectral6))
p = figure(x_range=fruits, plot_height=250, title="Fruit Counts",
           toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
show(p)

```

```python
#指定輸出檔名
output_file("fruit_info2.html")

#給定資料
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]
#資料做排序
sorted_fruits = sorted(fruits, key=lambda x: counts[fruits.index(x)])

# 不增設按鈕 toolbar_location=None, tools=""
# plot_height 設定條形高度
# 配置圖表
p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)


#Y座標從0開始
p.y_range.start = 0

show(p)
```
