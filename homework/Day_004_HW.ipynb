### 2. 如何在不用迴圈的情況下計算 (A+B)*(-A/2) ？那用迴圈怎麼做？


````python
A = np.ones(3)*1
B = np.ones(3)*2


print(A)
print(B)
````

````python
# 不用迴圈
print((A + B)*(-A/2))
````

````python
result=[]
for i in np.arange(A.size):
    result.append((A[i] + B[i])*(-A[i]/2))
````
