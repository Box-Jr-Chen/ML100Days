### Q1: 大樂透的頭獎，你必須從49個挑選出 6 個號碼，
### 且這六個號碼與頭獎的六個號碼一致，頭獎的機率是屬於哪一種分配?

### Ans : 超幾何分配(Hypergeometric Distribution)

### Q2: 運用範例的 python 程式碼，計算大樂透的中頭獎機率?

```python
'''
X: HG(49,6,6)
p(X=6) 的機率就是中頭獎的機率
'''
probs = stats.hypergeom.pmf(6,49,6,6)
print("中頭獎的機率為==",probs)
```

```python
N=49
K=6
n=6


#產生一個同樣間隔的序列
r = 6 #產出 x 對應點

# 2.計算機率質量函數 (probability mass function)
# P(X=x) --> 是機率
probs = stats.hypergeom.pmf(r,N,K,n)
print('------------')
print(probs)

```
