### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
A = np.random.random((5, 5))
A
````

````python
def normalization(data):
    _range = np.max(data) - np.min(data)
    return (data - np.min(data)) / _range
````

````python
normalization(A)
````
