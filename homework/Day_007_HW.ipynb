### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
Z.shape
```

```python
#直接運算
%timeit -n 10 a = 2

Z_1=Z+1
```

```python
#攤平運算後重新塑形
%timeit -n 10 a = 2
Z=(Z.flat+np.array([1])).reshape(100,100,100)
```

```python
#迭代物件運算後重新塑形
%timeit -n 10 a = 2

Z=np.array([i+1 for i in np.nditer(Z)]).reshape(100,100,100)
Z
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python

Z = np.random.randint(0,10,50)
print(Z)
```

```python
Z_max=[i for i in np.where(np.bincount(Z)==np.bincount(Z).max())[0]]
```

```python
Z_position=np.where(Z==Z_max)[0].tolist() 
```

```python

print("結果 最多的值是:"+str(Z_max) +"; 位置:"+str(Z_position))
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
```
